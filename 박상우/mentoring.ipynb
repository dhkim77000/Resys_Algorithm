```python
import numpy as np

# 랜덤 시드 설정
np.random.seed(42)

# 좌표 생성 함수
def generate_coordinates(shape, num_samples):
    if shape == 'cylinder':
        # cylinder 좌표 생성 (높이, 반지름, 각도)
        height = np.random.rand(num_samples) * 10 # 높이 범위: 0부터 10
        radius = np.random.rand(num_samples) * 5  # 반지름 범위: 0부터 5
        angle = np.random.rand(num_samples) * 2 * np.pi  # 각도 범위: 0부터 2*pi
        x = radius * np.cos(angle)
        y = radius * np.sin(angle)
        z = height
        return np.column_stack((x, y, z))
    elif shape == 'torus':
        # torus 좌표 생성 (반지름, 내부 반지름, 각도1, 각도2)
        radius = np.random.rand(num_samples) * 5 + 5  # 반지름 범위: 5부터 10
        inner_radius = np.random.rand(num_samples) * 2 + 1  # 내부 반지름 범위: 1부터 3
        angle1 = np.random.rand(num_samples) * 2 * np.pi  # 각도1 범위: 0부터 2*pi
        angle2 = np.random.rand(num_samples) * 2 * np.pi  # 각도2 범위: 0부터 2*pi
        x = (radius + inner_radius * np.cos(angle1)) * np.cos(angle2)
        y = (radius + inner_radius * np.cos(angle1)) * np.sin(angle2)
        z = inner_radius * np.sin(angle1)
        return np.column_stack((x, y, z))
    elif shape == 'sphere':
        # sphere 좌표 생성 (반지름, 각도1, 각도2)
        radius = np.random.rand(num_samples) * 5 + 5  # 반지름 범위: 5부터 10
        angle1 = np.random.rand(num_samples) * np.pi  # 각도1 범위: 0부터 pi
        angle2 = np.random.rand(num_samples) * 2 * np.pi  # 각도2 범위: 0부터 2*pi
        x = radius * np.sin(angle1) * np.cos(angle2)
        y = radius * np.sin(angle1) * np.sin(angle2)
        z = radius * np.cos(angle1)
        return np.column_stack((x, y, z))
    else:
        raise ValueError("Invalid shape. Choose from 'cylinder', 'torus', or 'sphere'.")

# 좌표 생성
cylinder_coordinates = generate_coordinates('cylinder', 10000)
torus_coordinates = generate_coordinates('torus', 10000)
sphere_coordinates = generate_coordinates('sphere', 10000)
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.manifold import Isomap

# 원본 좌표
coordinates = np.vstack((cylinder_coordinates, torus_coordinates, sphere_coordinates))

# PCA 차원 축소
pca = PCA(n_components=2)
pca_result = pca.fit_transform(coordinates)

# t-SNE 차원 축소
tsne = TSNE(n_components=2, random_state=42)
tsne_result = tsne.fit_transform(coordinates)

# ISOMAP 차원 축소
isomap = Isomap(n_components=2, n_neighbors=30)
isomap_result = isomap.fit_transform(coordinates)

# 차원 축소 결과 시각화
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
sns.scatterplot(x=pca_result[:, 0], y=pca_result[:, 1], s=1)
plt.title('PCA')
plt.xlabel('PC1')
plt.ylabel('PC2')

plt.subplot(1, 3, 2)
sns.scatterplot(x=tsne_result[:, 0], y=tsne_result[:, 1], s=1)
plt.title('t-SNE')
plt.xlabel('t-SNE 1')
plt.ylabel('t-SNE 2')

plt.subplot(1, 3, 3)
sns.scatterplot(x=isomap_result[:, 0], y=isomap_result[:, 1], s=1)
plt.title('ISOMAP')
plt.xlabel('ISOMAP 1')
plt.ylabel('ISOMAP 2')

plt.tight_layout()
plt.show()
```
