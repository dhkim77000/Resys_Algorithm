## 아이디어
* 스택이 비어있을 경우, 현재 값이 최대가 되며 스택에 현재 값을 저장한다.
* 현재 값이 스택에 저장된 최대값보다 작은 값일 경우 스택에 현재 값을 추가한다. 그렇지 않다면 스택의 길이를 출력한 뒤, 스택을 모두 비우고 현재 값을 저장한다.

```python
def solution(progresses, speeds):
    answer=[]
    stack=[]
    for i in range(len(progresses)):
        a,b = divmod((100-progresses[i]), speeds[i])
        s = a+bool(b)
        if stack and s>stack[0]:
            answer.append(len(stack))
            stack=[]
        stack.append(s)

#         print(stack)
    answer.append(len(stack))
    return answer
```

```python
progresses=[93, 30, 55]

speeds=[1,30,5]
solution(progresses,speeds)
```

```python
def solution(progresses, speeds):
    Q=[]
    for p, s in zip(progresses, speeds):
        if len(Q)==0 or Q[-1][0]<-((p-100)//s):
            Q.append([-((p-100)//s),1])
        else:
            Q[-1][1]+=1
    return [q[1] for q in Q]
```
