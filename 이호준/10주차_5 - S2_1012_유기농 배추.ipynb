```python
for _ in range(int(input())):
    m,n,k = map(int,input().split())
    land = [[0]*m for i in range(n)]
    for __ in range(k):
        x,y = map(int, input().split())
        land[y][x]=1
    cnt = 0
    for i in range(n):
        for j in range(m):
            if land[i][j]==1:
                cnt+=1
                land[i][j]=0
                work=[(i,j)]
                while work:
                    x,y = work.pop()
                    for x2,y2 in [(x-1,y),(x+1,y),(x,y-1),(x,y+1)]:
                        if 0<= x2 < n and 0<= y2 < m and land[x2][y2]:
                            land[x2][y2] = 0
                            work.append([x2,y2])
    print(cnt)
```
