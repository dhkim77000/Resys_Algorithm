## 아이디어
* 항상 주식의 가격이 떨어지지 않았다면, return은 [len(prices)-1, ... ,2,1,0] 의 형태를 갖는다.
* n초 이후, n초 시점의 가격보다 떨어진다면 해당 부분만 수정한다.


* 연산을 수행할 스택과, 답을 작성할 리스트 answer를 각각 한 개씩 생성한다. 
* 전과 비교했을때, 주식의 가격이 떨어지지 않았다면 스택에 현재 시점의 정보를 추가한다.
* 만약 주식 가격이 떨어졌다면, 현재 시점보다 높은 가격을 가진 과거 시점에 대한 정보를 모두 지운다. 이때, 스택에서 탈락한 과거시점들은 몇초간 가격을 유지하였는지 answer에 업데이트한다.

```python
def solution(prices):
    answer = list(range(len(prices)-1,-1,-1))
    stack = []
    for i,v in enumerate(prices):
        while stack and stack[-1][1]>v:
            answer[stack.pop()[0]]-=len(prices)-1-i
        stack.append([i,v])
#         print(stack)
    return answer
```

```python
prices = [1, 2, 1, 2, 3]
solution(prices)
```

```python
prices = [1, 2, 3, 2, 3]
solution(prices)
```

고수풀이(별로 고수아님)

```python
def solution(prices):
    answer = [0] * len(prices)
    for i in range(len(prices)):
        for j in range(i+1, len(prices)):
            if prices[i] <= prices[j]:
                answer[i] += 1
            else:
                answer[i] += 1
                break
    return answer
```
