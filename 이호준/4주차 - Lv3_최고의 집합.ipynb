* 곱했을 때 최대의 값이 되려면 딱 중간정도의 값이 되어야 한다. ex) s=16, n=4, result =[4,4,4,4]
* s를 n으로 나눈 값으로 구성된 길이의 리스트가 정답임. 다만 원소들이 자연수여야 하므로, 몫과 나머지를 적절히 사용하여 조정한다.

```python
n = 3
s = 9
solution(n,s)
```

```python
def solution(n, s):
    if n>s:
        return [-1]
    a,b=divmod(s,n)
    return [a]*(n-b) + [a+1]*b
```
