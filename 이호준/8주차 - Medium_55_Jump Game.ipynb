```python
class Solution:
    def canJump(self, nums:int) -> bool:
        dp = [False]*len(nums)
        can = nums[0]
        for i in range(1, len(nums)):
            if can>0:
                can-=1
                dp[i]=True
            else:
                return False
                break
            can = max(nums[i],can)
        return True
```
