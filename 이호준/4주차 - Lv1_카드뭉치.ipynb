# 아이디어
* 사용한 단어는 card1과 2에서 제거
* 만약 필요한 단어가 card1과 card2 중 없다면 No를 리턴

```python
cards1 = ["i", "water", "drink"]
cards2 = ["want", "to"]
goal = ["i", "want", "to", "drink", "water"]
```

```python
def solution()
    for word in goal:
        if cards1 and word == cards1[0]:
            cards1.pop(0)
        elif cards2 and word == cards2[0]:
            cards2.pop(0)
        else:
            return "No"
    return "Yes"
```

```python
solution(cards1, cards2, goal)
```

고수(?) 풀이

```python
def solution(cards1, cards2, goal):
    idx1,idx2=0,0
    for word in goal:
        if len(cards1)>idx1 and cards1[idx1]==word:
            idx1+=1
        elif len(cards2)>idx2 and cards2[idx2]==word:
            idx2+=1
        else:
            return "No"
    return "Yes"
```
