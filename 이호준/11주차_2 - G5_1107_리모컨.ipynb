```python
cc=int(input())
rnums=set()
if input()!='0':
    rnums=set(input())
cnt=0
up, down = cc, cc
while up != 100 and down != 100:
    print(up, down)
    if not set(str(down)) & rnums or not set(str(up)) & rnums:
        break
    if down>0:
        down-=1
    up+=1
    cnt+=1
cnt_a=[abs(100-cc)]
if not set(str(down)) & rnums:
    cnt_a.append(cnt+len(str(down)))
if not set(str(up)) & rnums:
    cnt_a.append(cnt+len(str(up)))
if up == 100 or down == 100:
    cnt_a.append(cnt)
print(min(cnt_a))
```
