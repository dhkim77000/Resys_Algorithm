## 아이디어
* 문서는 중요도가 높은 순서대로 처리된다.
* 중요도가 높은 순으로 정렬된 리스트 A를 생성한다.
* 주어진 priorities에 인덱스정보를 추가한 새로운 리스트(enumerate를 사용) B를 만든다.
* B를 계속 순서대로 순회하면서, 리스트 B에서 높은 중요도를 가진 문서순(리스트 A를 참고하여)으로 제거한다. 목표의 문서(location)가 처리될 때, 프로그램을 종료한다.

```python
def solution(priorities, location):
    sorted_p = sorted(priorities, reverse=True)
    p = list(enumerate(priorities))

    i=0
    for index, value in enumerate(sorted_p):
        while 1:
            i %= len(p)
#             print(index, value, i, p)
            if p[i][1]==value:
                if p.pop(i)[0]==location:
                    return index+1
                break
            i += 1
        
```

```python
p= [1, 1, 9, 1, 1, 1]
l = 0
solution(p,l)
```

```python
def solution(priorities, location):
    queue =  [(i,p) for i,p in enumerate(priorities)]
    answer = 0
    while True:
        cur = queue.pop(0)
        if any(cur[1] < q[1] for q in queue):
            queue.append(cur)
        else:
            answer += 1
            if cur[0] == location:
                return answer
```
